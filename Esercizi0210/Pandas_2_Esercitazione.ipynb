# Consegna: hai a disposizione 2 dataset con varie informazioni su alcuni giocatori. Trova il rating medio dei giocatori che hanno Attempts < 600

```python
import pandas as pd

file_path = './dataset/'
file_name = 'players_new_1.csv'

df = pd.read_csv(file_path + file_name, sep=',')
print(len(df))

df
```

## Il dataset non ha NaN, ne inserisco uno per testare il codice

```python
import numpy as np

df['Comp%'][3] = np.nan
df['Ints'][6] = np.nan
df['Comp%'][10] = np.nan

df
```

```python
for colname in df.columns:
    df[colname][8] = np.nan

df
```

## Cancello o sostituisco i NaN

```python
df_NaNtoZero = df.fillna(0)
df_NaNtoZero
```

```python
df = df.dropna(axis=0, how='all')
# Cancella le righe che hanno tutti i valori NaN
df
```

```python
df = df.dropna(axis=0)
# Cancella le righe che hanno almeno un valore NaN
df
```

## Dopo qualunque rimozione di una o più righe devo resettare l'indice

```python
df.reset_index(drop=True, inplace=True)
df
```

# Merge/Join

```python
file_name = 'players_new_2.csv'

df2 = pd.read_csv(file_path + file_name, sep=',')
len(df2)
df2
```

## Provo a fare il merge

```python
df_merge = pd.merge(df, df2) # se non specifico nulla, fa il merge con la colonna in comune
df_merge
```

### Cancello le colonne che non mi interessano

```python
df_merge.drop(columns=['Nome', 'Cognome'], inplace=True)
df_merge
```

### Faccio un filtro con min_attempts < 600

```python
min_attempts = 600
df_filtered = df_merge[df_merge['Attempts'] < min_attempts]
df_filtered
```

### Calcolo il rating medio

```python
media_rating = df_filtered['Rating'].mean()
print(f'Il rating medio dei giocatori con Attempts < {min_attempts} è: {media_rating:.2f}')
```
